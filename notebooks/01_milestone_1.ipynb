# Missing Values

### Identify Missing Values

```python
def make_data_dict(df):
    rows = []
    for c in df.columns:
        dtype = str(df[c].dtype)
        n_missing = df[c].isna().sum()
        pct_missing = n_missing / len(df) * 100
        sample = df[c].dropna().iloc[0] if df[c].dropna().size>0 else None
        rows.append((c, dtype, n_missing, pct_missing, sample))
    return pd.DataFrame(rows, columns=['column','dtype' ,'n_missing' ,'pct_missing','example_sample'])

data_dict = make_data_dict(df)
data_dict.to_csv('../reports/data_dictionary.csv', index=False)
```

```python
data_dict.head(5)
```

```python
data_dict.sample(n=5)
```

```python
missing_cols = data_dict[data_dict['n_missing'] > 0]['column'].tolist()
print("Columns with missing values:", missing_cols)
df['condition_text'].value_counts()
```

### Handling Missing Values

```python
df['condition_text'] = df['condition_text'].fillna('Unknown')
numeric_cols = ['temperature_celsius', 'temperature_fahrenheit',
    'wind_mph', 'wind_kph', 'wind_degree',
    'pressure_mb', 'pressure_in',
    'precip_mm', 'precip_in',
    'humidity']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

df['last_updated'] = pd.to_datetime(df['last_updated'], format='%d-%m-%Y %H:%M')
```

```python
df = df.drop_duplicates()
```

# Anomalies - assess the quality and consistency of data

### Identify anomalies - numerical

```python
df_clean = df.copy()
```

```python
numeric_cols = [
    'temperature_celsius', 'temperature_fahrenheit', 
    'wind_mph', 'wind_kph', 'wind_degree', 
    'pressure_mb', 'pressure_in', 
    'precip_mm', 'precip_in', 
    'humidity', 'cloud', 
    'feels_like_celsius', 'feels_like_fahrenheit',
    'visibility_km', 'visibility_miles', 
    'uv_index', 'gust_mph', 'gust_kph', 
    'air_quality_Carbon_Monoxide', 'air_quality_Ozone', 
    'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide', 
    'air_quality_PM2.5', 'air_quality_PM10', 
    'air_quality_us-epa-index', 'air_quality_gb-defra-index', 
    'moon_illumination'
]

df[numeric_cols].describe().to_csv('../reports/numeric_cols_full_description.csv', index=False)
df[numeric_cols].describe()
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df[numeric_cols]))
anomalies = (z_scores > 3).any(axis=1) 
df_anomalies = df[anomalies]
df_anomalies.to_csv('../reports/numeric_cols_zscore_anomalies.csv', index=False)
df_anomalies
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numeric_cols])
plt.xticks(rotation=45)
plt.title("Box Plot of Numeric Columns")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['temperature_celsius'], df['humidity'], label='Normal', alpha=0.5)
plt.scatter(df_anomalies['temperature_celsius'], df_anomalies['humidity'], color='red', label='Anomalies')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.title('Temperature vs Humidity with Anomalies Highlighted')
plt.legend()
plt.show()
```

```python
sns.heatmap(z_scores, cmap='coolwarm', cbar=True)
plt.title("Z-Score Heatmap of Numeric Columns")
plt.show()
```

# Normalize Values

### StandardScaler (Z-score normalization)

```python
modified_numeric_cols = [
    'temperature_celsius', 
    'wind_mph',  'wind_degree', 
    'pressure_mb',  
    'precip_mm',  
    'humidity', 'cloud', 
    'feels_like_celsius', 
    'visibility_km',  
    'uv_index', 'gust_mph',  
    'air_quality_Carbon_Monoxide', 'air_quality_Ozone', 
    'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide', 
    'air_quality_PM2.5', 'air_quality_PM10', 
    'air_quality_us-epa-index', 'air_quality_gb-defra-index', 
    'moon_illumination'
]
```

###  StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_array = scaler.fit_transform(df_clean[modified_numeric_cols]) 

df_scaled = pd.DataFrame(scaled_array, columns=modified_numeric_cols, index=df_clean.index)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Original
original = df_clean[modified_numeric_cols]

# Standard Scaled
standard_scaled = pd.DataFrame(StandardScaler().fit_transform(original), 
                               columns=modified_numeric_cols, index=df_clean.index)
```

```python
features_to_plot = ['temperature_celsius', 'humidity', 'uv_index']

for feature in features_to_plot:
    plt.figure(figsize=(15, 4))

    plt.subplot(1, 3, 1)
    sns.histplot(original[feature], kde=True, color='gray')
    plt.title(f'Original: {feature}')

    plt.subplot(1, 3, 2)
    sns.histplot(standard_scaled[feature], kde=True, color='blue')
    plt.title(f'Standard Scaled: {feature}')


    plt.tight_layout()
    plt.show()
```

```python
for feature in features_to_plot:
    melted = pd.DataFrame({
        'Original': original[feature],
        'Standard': standard_scaled[feature]
    }).melt(var_name='Scaling', value_name='Value')
    
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Scaling', y='Value',hue='Scaling', data=melted, palette=['gray', 'blue'], legend=False)
    plt.title(f'Boxplot Comparison: {feature}')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
for feature in features_to_plot:
    plt.figure(figsize=(12, 4))
    plt.plot(original[feature].values[:100], label='Original', color='gray')
    plt.plot(standard_scaled[feature].values[:100], label='Standard', color='blue')
    plt.title(f'Row-wise Comparison: {feature} (First 100 rows)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# FINAL DATA

```python
non_numeric = df_clean.drop(columns=modified_numeric_cols)
df_final = pd.concat([standard_scaled, non_numeric], axis=1)
```

### Aggregate to Monthly Averages

```python
# Convert timestamp column
df_final['last_updated'] = pd.to_datetime(df_final['last_updated'])

# Extract month period
df_final['month'] = df_final['last_updated'].dt.to_period('M')

# Group by month and location
df_monthly = df_final.groupby(['month', 'country', 'location_name', 'latitude', 'longitude']).mean(numeric_only=True).reset_index()
```

```python
df_monthly.to_csv('../data/processed/final_weather_monthly.csv', index=False)
```
